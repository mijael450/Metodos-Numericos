```python
import numpy as np
import matplotlib.pyplot as plt 
import math

# Función para la cual se aplicará el método de bisección
def f(x):
    return x**3 - 7*x**2 + 14*x - 6

# Método de Bisección
def bisection_method(func, a, b, tol=1e-2, max_iterations=100):
    if func(a) * func(b) >= 0:
        print("El método de bisección no es aplicable en este intervalo.")
        return None
    
    iteration = 0
    while (b - a) / 2 > tol and iteration < max_iterations:
        midpoint = (a + b) / 2
        if func(midpoint) == 0:
            return midpoint  
        elif func(a) * func(midpoint) < 0:
            b = midpoint
        else:
            a = midpoint
        iteration += 1
    print(iteration)
    return (a + b) / 2  

a, b = 0, 1

root = bisection_method(f, a, b)
x_values = np.linspace(a, b, 500)
y_values = f(x_values)

plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label="f(x) = x^3 - 7x^2 + 14x - 6", color='blue')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(root, color='red', linestyle="--", label=f"Raiz ≈ {root:.2f}")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Gráfica de la función ")
plt.legend()
plt.grid(False)
plt.show()
```

```python
import math
from scipy.optimize import fsolve

# Datos dados
L = 10      # Longitud en cm
r = 1       # Radio en cm
V = 12.4    # Volumen en cm^3

# Definimos la función para el volumen en función de h
def volumen_func(h):
    # Calcula el volumen basado en la fórmula dada
    term1 = 0.5 * math.pi * r**2
    term2 = r**2 * math.asin(h / r)
    term3 = h * math.sqrt(r**2 - h**2)
    volumen_calculado = L * (term1 - term2 - term3)
    
    # La función devuelve la diferencia entre el volumen deseado y el calculado
    return volumen_calculado - V

# Estimación inicial para h (por ejemplo, la mitad del radio)
h_inicial = r / 2

# Usamos fsolve para encontrar el valor de h
h_solucion, = fsolve(volumen_func, h_inicial)

# Mostramos el resultado
print(f"La profundidad del agua en el abrevadero es aproximadamente {h_solucion:.4f} cm")
```

```python
import numpy as np
import matplotlib.pyplot as plt


a, b = 1, 2
epsilon = 1e-4

N = np.ceil((np.log(b - a) - np.log(epsilon)) / np.log(2))
print("Número mínimo de iteraciones necesarias:", int(N))


def f(x):
    return x**3 - x - 1


def bisection_method_precise(func, a, b, tol=epsilon, max_iterations=100):
    if func(a) * func(b) >= 0:
        print("El método de bisección no es aplicable en este intervalo.")
        return None, []
    
    iteration = 0
    midpoints = []  
    while (b - a) / 2 > tol and iteration < max_iterations:
        midpoint = (a + b) / 2
        midpoints.append(midpoint)  
        if func(midpoint) == 0:
            return midpoint, midpoints 
        elif func(a) * func(midpoint) < 0:
            b = midpoint
        else:
            a = midpoint
        iteration += 1
    
    return (a + b) / 2, midpoints 

root_precise, midpoints = bisection_method_precise(f, a, b)
print("Aproximación de la raíz:", root_precise)


x = np.linspace(0, 2, 100)  
y = f(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='f(x) = x³ - x - 1', color='blue')
plt.axhline(0, color='black', linewidth=0.5, ls='--')  
plt.axvline(root_precise, color='red', linestyle='--', label='Raíz aproximada')


plt.title('Método de Bisección')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid()
plt.ylim(-3, 3)  
plt.show()
```
